```python
# Base repo bath for files.
base_path_win = r'C:\Users\rsorma\Desktop\brazilian-ecommerce'
base_path_mac = r'/Users/roccosorma/Desktop/brazilian-ecommerce'

base_file_path = base_path_mac if comp_plat == 'Darwin' else base_path_win

# List of all files in the repo.
base_files = glob.glob(os.path.join(base_file_path, "*.csv"))
# Print desired file name split and dataframe variable assignment
#in addtion to creating a list of dataframe names that will gather
#all column names for id optimization.
list_of_df_names = []
list_of_df_col_names = []
list_of_df = []
for i in base_files:
    df = pacsv.read_csv(i).to_pandas()
    list_of_df.append(df)
    list_of_df_col_names.append(list(df.columns))
    list_of_df_names.append(str(i.split('/')[-1].replace('.csv', '') + '_df'))
```
